31. What was the most ordered item in the choice_description column?

```python

df.groupby('choice_description')['quantity'].sum().sort_values(ascending=False).idxmax()
```

32. What is the average revenue amount per order?

```python
# (df.quantity * df.item_price)
revenue = (df.quantity * df.item_price)
df.groupby(['order_id'])[revenue].mean()
```

33. Which product has the highest total quantity sold?

```python
ans = df.groupby('item_name')['quantity'].sum().sort_values(ascending=False)
ans.head(1)
```

34. calculate the average price of each product.

```python
df.groupby('item_name')['item_price'].mean()
```

35. Which choice appears most often?

```python
df['item_name'].value_counts().sort_values(ascending=False).head(1)
```
